##### Задание 1.2) Напишите код на Python, реализующий расчет длины вектора, заданного его координатами

```python
v1 = [10, 10, 10]             # координаты 1-го вектора
v2 = [0, 0, -10]             # координаты 2-го вектора
```

```python
def sum_of_squares(v):
    return sum(vi ** 2 for vi in v)
```

```python
def length(v):
    return math.sqrt(sum_of_squares(v))
```

```python
length(v1)           # длина первого вектора
```

```python
length(v2)           # длина второго вектора
```

##### 3.3.1 Напишите код, который будет переводить полярные координаты в декартовы

```python
def pol2cart(radius, alfa):
    x = radius * np.cos(alfa)
    y = radius * np.sin(alfa)
    return(x, y)
```

##### 3.3.3 Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах

```python
with plot.style.context("seaborn-white"):
    fig = plt.figure(figsize=(5,5))
    ax = fig.add_subplot(111, projection="polar")
    # три линии координат
    for degree in [90, 210, 330]:
        rad = np.deg2rad(degree)
        ax.plot([rad,rad], [0,1], color="black", linewidth=2)
    # Отрезок прямой линии ()
    for curve in [[[90, 210], [0.5, 0.8]]]:
        curve[0] = np.deg2rad(curve[0])
        ax.plot(curve[0], curve[1])
```

##### 3.4.1) Решите систему уравнений 

```python
# y = x**2 - 1
# exp(x) + x * (1 - y) = 1
```

```python
x = np.linspace(-3, 6, 201)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-5, 40) 
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)
```

```python
x1, y1 =  fsolve(equations, (-2, 3))          # 1-я пара корней системы уравнений
print (x1, y1)
```

```python
x2, y2 =  fsolve(equations, (2, 5))        # 2-я пара корней системы уравнений
print (x2, y2)
```

```python
x3, y3 =  fsolve(equations, (5, 25))       # 3-я пара корней системы уравнений
print (x3, y3)
```
