#### Задание 17.6.6

##### 3*x**2 + 5*y**2 + 12*x - 30*y + 42 = 0

Собираем:

3*(x+1)**2 - 3 + 5*(y-3)**2 - 45 + 42 = 0

3*(x+1)**2 + 5*(y-3)**2 = 6

1/2*((x+1)**2) + 5/6*(y-3)**2 = 1

##### Ответ: данная кривая - эллипс, с центром эллипса в точке координат (-1; 3)

```python
x = np.linspace(-9, 9, 300)
y = np.linspace(-5, 5, 300)
x, y = np.meshgrid(x, y)

plt.axis([-2, 0, 1, 5])

a = 1/2
b = 5/6
plt.contour(x, y,((x+1)**2/a**2 + (y-3)**2/b**2), [1], colors='k')

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
